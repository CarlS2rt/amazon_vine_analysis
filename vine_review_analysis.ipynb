```python
import pandas as pd
```

```python
file = "vine_table.csv"
vine = pd.read_csv(file, parse_dates=True)
```

```python
vine.head()
```

```python
vine_filter = vine.loc[vine['total_votes'] >= 20]
vine_filter.head()
```

```python
vine_helpful = vine_filter.loc[vine_filter['helpful_votes']/vine_filter['total_votes'] >= 0.50]
vine_helpful
```

```python
vine_paid = vine_helpful.loc[vine_helpful['vine'] == 'Y']
vine_paid.head()
```

```python
vine_unpaid = vine_helpful.loc[vine_helpful['vine'] == 'N']
vine_unpaid.head()
```

```python
total_reviews = vine_helpful['review_id'].count()
total_paid_reviews = vine_paid['review_id'].count()
total_unpaid_reviews = vine_unpaid['review_id'].count()
```

```python
total_5star = vine_helpful['star_rating'].loc[vine_helpful['star_rating'] == 5].count()
```

```python
paid_5star = vine_paid['star_rating'].loc[vine_paid['star_rating'] == 5].count()
```

```python
unpaid_5star = vine_unpaid['star_rating'].loc[vine_unpaid['star_rating'] == 5].count()
```

```python
total_percent_5star = total_5star / total_reviews * 100
```

```python
paid_percent_5star = paid_5star / total_paid_reviews * 100
```

```python
unpaid_percent_5star = unpaid_5star / total_unpaid_reviews * 100
```

```python
print(f'total reviews {total_reviews}, total paid reviews {total_paid_reviews}, total unpaid reviews {total_unpaid_reviews}')
print(f'total 5-star reviews {total_5star}, total paid 5-star {paid_5star}, total unpaid 5-star {unpaid_5star}')
print(f'total % 5-star reviews {total_percent_5star}, total % paid 5-star reviews {paid_percent_5star}, total % unpaid 5-star reviews {unpaid_percent_5star}')
```
